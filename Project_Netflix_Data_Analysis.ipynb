Plotting a Pie Chart using matplotlib.pyplot for displaying the share of TV Show and Movie

```python
data['type'].value_counts().plot.pie(autopct='%1.1f%%')
```

Plotting a Bar Chart between year of release and views

```python
data.head(20).plot(kind='bar',x='release_year',y='total views')
```

Plotting a Pie Chart to visualise the categories of genres that are present

```python
data['listed_in'].value_counts().head(10).plot.pie(autopct='%1.1f%%')
```

Filling the null values in the listed_in column with the value TV-G

```python
data['listed_in'].fillna("TV-G")
```

Displaying the Pie Chart for the distribution of the Star Rating given by users

```python
data['star rating'].head(20).value_counts().plot.pie(autopct='%1.1f%%')
```

Using Subplot to plot two Charts

```python
plt.subplot(2,1,1)
plt.bar(data['q1 views'],data['q2 views'])
plt.xlabel('Q1 views')
plt.ylabel('Q2 views')
plt.title('Q1 vs Q2 & Q3 vs Q4')

plt.subplot(2,1,2)
plt.bar(data['q3 views'],data['q4 views'])
plt.xlabel('Q3 views')
plt.ylabel('Q4 views')
```

Using Bar Plot to compare Quarterly Views with Total Views

```python
data.head(10).plot(kind='bar',x='q1 views',y='total views')
data.head(20).plot(kind='bar',x='q2 views',y='total views')
data.head(20).plot(kind='bar',x='q3 views',y='total views')
data.head(20).plot(kind='bar',x='q4 views',y='total views')
```

Comparing number of TV Shows and Movies released in a year

```python
data.head(10).groupby(['release_year','type']).size().unstack().plot(kind='barh')
```

Plotting Horizontal Bar Chart between Quarterly Views

```python
data.head(10).groupby(['q1 views','q2 views','q3 views','q4 views']).size().unstack().plot(kind='barh')
```

Ploting a Bar Chart between Rating of Movies and TV Shows

```python
data.head(10).groupby(['star rating','type']).size().unstack().plot(kind='barh')
```

Using Bar Plot to compare Quarterly Views with Rewatched Views

```python
data.head(20).plot(kind='bar',x='q1 views',y='rewatched',color="g")
data.head(20).plot(kind='bar',x='q2 views',y='rewatched',color="r")
data.head(20).plot(kind='bar',x='q3 views',y='rewatched',color="b")
data.head(20).plot(kind='bar',x='q4 views',y='rewatched',color="grey")
```
